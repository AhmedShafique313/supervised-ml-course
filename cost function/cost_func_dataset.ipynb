```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('employee_salary.csv')
```

```python
x_train = dataset['Experience_Years'].values
y_train = dataset['Salary'].values
```

```python
def cost_func_compute(x,y,w,b):
    m = np.shape(x)[0]
    sum_of_cost = 0
    for i in range(m):
        f_wb = w*x[i] + b
        cost = (f_wb-y[i])**2
        sum_of_cost = sum_of_cost + cost
        total = (1/(2*m)*sum_of_cost)
    return total
```

```python
w_values = np.linspace(-50, 50, 50) 
cost_values = [cost_func_compute(x_train, y_train, w, 0) for w in w_values]
```

```python
plt.plot(w_values, cost_values, color='red')
plt.xlabel('w')
plt.ylabel('Cost')
plt.title('Cost Function')
plt.show()
```
